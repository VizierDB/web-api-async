```python
from vizier.datastore.histore.base import HistoreDatastore
TMP_DIR = '../.tmp'
```

```python
store = HistoreDatastore(TMP_DIR)
```

```python
from vizier.filestore.base import CSV

ds = store.load_dataset(CSV('../data/etnx-8aft.tsv'), profiler='datamartprofiler')
```

```python
import json
print(json.dumps(ds.profiling('datamartprofiler'), indent=4))
```

```python
df = ds.to_dataframe()
df.head()
```

```python
from openclean import op
from openclean.function import Sum

df = op.inscol(df, 'Abuse Total', pos=4, values=Sum('Physical Abuse', 'Sexual Abuse'))
ds = store.update_dataset(origin=ds, df=df, profiler='datamartprofiler')
df = store.get_dataset(ds.identifier).to_dataframe()
df
```

```python
print(json.dumps(ds.profiling('datamartprofiler'), indent=4))
```
